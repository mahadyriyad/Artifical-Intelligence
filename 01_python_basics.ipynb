```python
def calculate_area(length, width):
  """Calculates the area of a rectangle.

  Args:
    length: The length of the rectangle.
    width: The width of the rectangle.

  Returns:
    The area of the rectangle.
  """
  area = length * width
  return area

# Call the function with arguments
rectangle_area = calculate_area(5, 10)
print(f"The area of the rectangle is: {rectangle_area}")
```

### Function Example Details

This `calculate_area` function demonstrates:

*   **Definition**: It starts with `def` followed by the function name and parameters in parentheses.
*   **Parameters**: `length` and `width` are inputs the function takes.
*   **Docstring**: The triple-quoted string explains what the function does, its arguments (`Args`), and what it returns (`Returns`). This is good practice for documentation.
*   **Body**: The indented lines of code perform the calculation.
*   **Return Value**: The `return area` statement sends the calculated `area` back as the result of the function call.
*   **Calling the Function**: `calculate_area(5, 10)` executes the function with specific values for `length` and `width`.
