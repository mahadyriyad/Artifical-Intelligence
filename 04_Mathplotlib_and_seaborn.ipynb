```python
#Importing Matplotlib
import matplotlib.pyplot as plt
```

```python

#Simple Line Plot

x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

plt.plot(x, y)   # line plot
plt.show()
```

```python
plt.plot(x, y)
plt.xlabel("X-axis")       # X label
plt.ylabel("Y-axis")       # Y label
plt.title("Simple Line Graph")  # Title
plt.show()
```

```python
#Customizing Line
plt.plot(x, y, color="red", marker="s", linestyle="--", linewidth=2)
plt.show()
```

```python
#Multiple Lines

y2 = [1, 4, 9, 16, 25]
plt.plot(x, y, label="y=2x")
plt.plot(x, y2, label="y=x²")
plt.legend()   # show labels
plt.show()
```

```python
plt.scatter(x, y, color='red')
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Scatter Plot")
plt.show()
```

```python
categories = ["A", "B", "C", "D", "E"]
values = [4, 7, 2, 8, 6]

plt.bar(categories, values, color='orange')
plt.xlabel("Categories")
plt.ylabel("Values")
plt.title("Bar Chart")
plt.show()
```

```python
#Pie Chart
sizes = [20, 30, 25, 25]
labels = ["A", "B", "C", "D"]

plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title("Pie Chart")
plt.show()
```

```python
# Example dataset
data = [7, 8, 5, 6, 9, 12, 15, 14, 6, 8, 7, 30]  # outlier = 30

plt.boxplot(data)
plt.title("Box Plot (Outlier Detection)")
plt.ylabel("Value")
plt.show()
```

```python
#Multiple Plots → plt.subplot()
 #Use plt.subplot(rows, cols, index)

import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y1 = [1, 4, 9, 16, 25]
y2 = [2, 4, 6, 8, 10]

# Create 1 row, 2 columns → plot 1
plt.subplot(1, 2, 1)
plt.plot(x, y1, 'r-o')
plt.title("y = x²")
plt.grid(True)

# plot 2
plt.subplot(1, 2, 2)
plt.plot(x, y2, 'b--s')
plt.title("y = 2x")
plt.grid(True)

plt.show()
```

```python
import pandas as pd

#heatmap
data = {
    "Math":    [85, 90, 78, 92, 88],
    "Physics": [80, 85, 75, 89, 90],
    "Chem":    [78, 88, 82, 85, 87],
    "Biology": [92, 94, 88, 96, 90]
}
df = pd.DataFrame(data)
corr = df.corr()

plt.imshow(corr, cmap="coolwarm", interpolation="nearest")
plt.colorbar()
```

# Seaborn

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load sample dataset
data = sns.load_dataset("tips")

# Simple plot
sns.scatterplot(x="total_bill", y="tip", data=data)
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.barplot(x='day', y='total_bill', data=data)
plt.title('Average Total Bill by Day')
plt.xlabel('Day of the Week')
plt.ylabel('Average Total Bill')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(data=data, x='total_bill', bins=15, kde=True)
plt.title('Distribution of Total Bill')
plt.xlabel('Total Bill')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.boxplot(x='day', y='total_bill', data=data)
plt.title('Total Bill Distribution by Day')
plt.xlabel('Day of the Week')
plt.ylabel('Total Bill')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.violinplot(x='day', y='total_bill', data=data)
plt.title('Total Bill Distribution by Day (Violin Plot)')
plt.xlabel('Day of the Week')
plt.ylabel('Total Bill')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.countplot(x='day', data=data)
plt.title('Count of Tips by Day')
plt.xlabel('Day of the Week')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.kdeplot(x='tip', data=data, fill=True, color='purple')
plt.title('Kernel Density Estimate of Tip Amounts')
plt.xlabel('Tip Amount')
plt.ylabel('Density')
plt.show()
```

# Scatter Plots

Scatter plots are used to visualize the relationship between two numerical variables. Each point on the plot represents an observation, and its position is determined by the values of the two variables.

```python
x = [11,12,3,22]
y= [5,9,11,15]
plt.scatter(x,y)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

# Subplot

```python
x = [1,3,2,5]
plt.figure(figsize=(10,5))

#subplot-1
plt.subplot(1,2,1)
plt.plot(x,[i**2 for i in x])
plt.title("Square")

#subplot-2
plt.subplot(1,2,2)
plt.bar(x,[i**3 for i in x])
plt.title("Cube")

#subplot-3
plt.subplot(2,2,3)
plt.scatter(x,[i**3 for i in x])
plt.title("Square")

plt.tight_layout()
plt.show()
```

```python
import pandas as pd

# Define the data and create the DataFrame
data = {
    "Math":    [85, 90, 78, 92, 88],
    "Physics": [80, 85, 75, 89, 90],
    "Chem":    [78, 88, 82, 85, 87],
    "Biology": [92, 94, 88, 96, 90]
}
df = pd.DataFrame(data)

# Calculate the correlation matrix
corr = df.corr()

plt.figure(figsize=(10,5))
sns.heatmap(corr, cmap="coolwarm", annot=True , linewidths=0.5)
plt.show()
```
