First, let's confirm the missing values in the `titanic` dataset.

```python
print('Missing values before handling:')
df.isnull().sum()
```

```python
df = pd.read_csv('/content/titanic.csv')

# Fill missing 'Age' values with the mean age
df['Age'].fillna(df['Age'].mean(), inplace=True)

# Fill missing 'Fare' values with the mean fare
df['Fare'].fillna(df['Fare'].mean(), inplace=True)

# Drop the 'Cabin' column due to many missing values
df.drop('Cabin', axis=1, inplace=True)

print('\nMissing values after handling:')
display(df.isnull().sum())
```

```python
display(df.head())
```

Now, let's handle the missing values:

*   **Age**: Fill missing values with the mean age.
*   **Fare**: Fill missing values with the mean fare.
*   **Cabin**: This column has a large number of missing values, so we'll drop it.

```python
df = pd.read_csv('/content/titanic.csv')

# Fill missing 'Age' values with the mean age
df['Age'].fillna(df['Age'].mean(), inplace=True)

# Fill missing 'Fare' values with the mean fare
df['Fare'].fillna(df['Fare'].mean(), inplace=True)

# Drop the 'Cabin' column due to many missing values
df.drop('Cabin', axis=1, inplace=True)

print('\nMissing values after handling:')
display(df.isnull().sum())
```

# Group by

```python
data = {
    "city":["Dhaka" , "Dhaka" , "Cumilla" , "Cumilla"],
    "salary": [1000,3000,600,4000]
    }

df = pd.DataFrame(data)

grouped = df.groupby("city")["salary"].sum()
print(grouped)
```
